**[Task 2.2.]()** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical $3 \times 3$ Sobel filtering, respectively:

```python
gv = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
gh = np.array([[-1,-2,-1],[0,0,0,],[1,2,1]])

lenav=np.zeros_like(lena)
lenah=np.zeros_like(lena)

for x in range (1, len(lena)-1):
    for y in range(1, len(lena)-1):
        local = lena[x-1:x+2,y-1:y+2]
        lenav[x,y]=np.sum(local * gv)

for x in range (1, len(lena)-1):
    for y in range(1, len(lena)-1):
        local = lena[x-1:x+2,y-1:y+2]
        lenah[x,y]=np.sum(local * gh)

plt.subplot(1,2,1)
plt.imshow(lenav,cmap="grey")

plt.subplot(1,2,2)
plt.imshow(lenah,cmap="grey")

```
