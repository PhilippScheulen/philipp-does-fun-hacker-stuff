```python
import numpy as np
import matplotlib.pyplot as plt
```

**[Task 1.2 (a).]()** Use `plt.imread` to load an image:

```python
image = plt.imread("data/NIH3T3/im/dna-0.png")
```

**[Task 1.2 (b).]()** Show the image:

```python
plt.figure(figsize=(13, 6))
plt.imshow (image, cmap="grey")
plt.colorbar()
```

**[Task 1.2 (d).]()** Binarize the image by applying an intensity threshold:

```python
threshold = 0.1

binary_image =(image >= threshold)
plt.figure(figsize=(15,8))
plt.imshow(binary_image, cmap="grey")
```

**[Task 1.2 (e).]()** Load the ground truth:

```python
truth = plt.imread("data/NIH3T3/gt/0.png")
plt.imshow(truth)
```

Compute and print the Dice coefficient:

```python
Dicer = 2*np.sum(binary_image * truth)/(binary_image.sum() + truth.sum())
print (Dicer)
```
