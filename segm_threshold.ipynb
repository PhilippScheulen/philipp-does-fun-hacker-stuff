**[Task 1.2 (a).]()** Use `plt.imread` to load an image:

**[Task 1.2 (b).]()** Show the image:

**[Task 1.2 (d).]()** Binarize the image by applying an intensity threshold:

```python
if len(image.shape) == 3:  # Check if it's an RGB image
    grayscale_image = np.mean(image, axis=-1)  # Average RGB channels
else:
    grayscale_image = image  # Image is already grayscale

threshold = 0.3

binary_image =(grayscale_image >= threshold).astype(np.uint8)
plt.figure(figsize=(15,8))

# Grayscale image
plt.subplot(1, 2, 1)
plt.imshow(grayscale_image, cmap="gray")
plt.title("Grayscale Image")
plt.colorbar()

# Binarized image
plt.subplot(1, 2, 2)
plt.imshow(binary_image, cmap="gray")
plt.title(f"Binarized Image (Threshold = {threshold})")
plt.colorbar()
plt.imshow(binary_image)
```
